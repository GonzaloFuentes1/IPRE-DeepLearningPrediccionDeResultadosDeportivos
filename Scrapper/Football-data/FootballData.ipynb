## Para ordenarlos 

```python
leagues = ['ARG-PrimeraDivision', 'BRA-Brasileirao','ENG-Premier League', 'ESP-La Liga', 'FRA-Ligue 1',
           'GER-Bundesliga', 'ITA-Serie A', 'NED-Eredivisie', 'POR-PrimeiraDivisao', 'RUS-PremierLeague', 'SCO-Premiership', 'TUR-SuperLig', 'USA-MLS']
# Chat GPT
def fix_date_format(date_str):
    # Divide la cadena de fecha en dÃ­a, mes y aÃ±o
    parts = date_str.split('/')
    
    # Formatea el aÃ±o a cuatro dÃ­gitos si es necesario
    year = parts[-1]
    if len(year) == 2:
        if int(year) > 50:
            year = f"19{year}"
        else:
            year = f"20{year}"
    
    # Formatea la fecha como XX/XX/XXXX
    formatted_date = f"{parts[0]}/{parts[1]}/{year}"
    return formatted_date

for l in leagues:
    df = pd.read_csv(f"FootballData_{l}.csv").dropna(subset=["Date"])
    df['Date'] = df['Date'].apply(fix_date_format)
    df = df.sort_values(by=["Date","Time"],ignore_index=True)
    df.to_csv(f"FootballData_{l}.csv",index=False)
```
