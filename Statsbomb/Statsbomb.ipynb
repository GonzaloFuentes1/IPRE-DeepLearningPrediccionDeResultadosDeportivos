```python
tiros = argentina_francia[argentina_francia["type"]== "Shot"]
pases = argentina_francia[argentina_francia["type"]== "Pass"]
```

```python
tiros[["x", "y"]]= tiros["location"].apply(pd.Series)
pases[["x", "y"]]= pases["location"].apply(pd.Series)
tiros[["shot_end_x", "shot_end_y","a"]]= tiros["shot_end_location"].apply(pd.Series)
pases[["pass_end_x", "pass_end_y"]]= pases["pass_end_location"].apply(pd.Series)
```

```python
tiros_argentina = tiros[tiros["team"] == "Argentina"]
tiros_francia = tiros[tiros["team"] == "France"]
pases_argentina = pases[pases["team"] == "Argentina"]
pases_francia = pases[pases["team"] == "France"]
```

```python
tiros_argentina.head()
```

```python
'''
Graficando pases en campo de futbol
'''
sbred = '#FF0000'
#detalles de los colores de las lineas en el grafico
cmaplist = ["white",'#00FFFF', '#00FF80','#00FF00','#80FF00','#FFFF00','#FF8000','#FF0000']
cmap = LinearSegmentedColormap.from_list("", cmaplist)
```

```python
def shotMap(tiros):
    # Cancha
    pitch = Pitch(pitch_type='statsbomb', pitch_color='white',
                  line_color='black', line_zorder=2)
    fig, ax = pitch.draw(
        figsize=(16, 11), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('white')

    # Tiros
    bin_statistic = pitch.bin_statistic(
        tiros.shot_end_x, tiros.shot_end_y, statistic="count", bins=(48, 24), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, alpha=1, cmap=cmap)

    # Flechas
    pitch.arrows(tiros.x, tiros.y, tiros.shot_end_x, tiros.shot_end_y, width=1,
                 headwidth=8, headlength=5, color=sbred, ax=ax, zorder=2, label="Tiros")
    shot_position = plt.scatter(tiros.x, tiros.y)

def passMap(pases):
    # Cancha
    pitch = Pitch(pitch_type='statsbomb', pitch_color='white',
                  line_color='black', line_zorder=2)
    fig, ax = pitch.draw(
        figsize=(16, 11), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('white')

    # Pases
    bin_statistic = pitch.bin_statistic(
        pases.pass_end_x, pases.pass_end_y, statistic="count", bins=(12, 8), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, alpha=1, cmap=cmap)

    # Flechas
    pitch.arrows(pases.x, pases.y, pases.pass_end_x, pases.pass_end_y, width=1,
                 headwidth=8, headlength=5, color=sbred, ax=ax, zorder=2, label="Pases")
    pases_position = plt.scatter(pases.x, pases.y)
    
```

```python
shotMap(tiros_francia)
```

```python
passMap(pases_argentina)
```

```python
passMap(pases_francia)
```
